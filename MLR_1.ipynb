```python
x=df.iloc[:,:-1] # x,y split
y=df.iloc[:,-1]
```

```python
from sklearn.preprocessing import LabelEncoder #Label encoding
lb=LabelEncoder()
x['Product_Name']=lb.fit_transform(x['Product_Name'])
df['Product_Name']=lb.fit_transform(x['Product_Name'])
```

```python
from sklearn.model_selection import train_test_split   # Train_Test_Split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=7,test_size=0.3)
```

```python
x_train.shape
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression  # Model building
s_regression=LinearRegression()
```

```python
s_regression.fit(x_train,y_train)
```

```python
Y_predict=s_regression.predict(x_test)  #Model evalution/model prediction
```

```python
Y_predict
```

```python
y_test
```

```python
from sklearn.metrics import (mean_absolute_error,mean_squared_error)
```

```python
MAE = mean_absolute_error(y_test,Y_predict)
```

```python
MAE
```

```python
MSE = mean_squared_error(y_test,Y_predict)
```
